```python
neutral_path ='/content/drive/MyDrive/졸프/neutral.csv'
emotional_path ='/content/drive/MyDrive/졸프/emotional.xlsx'

```

```python
import pandas as pd
```

```python
neutral_train_data = pd.read_csv(neutral_path)
emotional_train_data = pd.read_excel(emotional_path)
```

```python
selected_columns = ["사람문장1", "감정_대분류"]
emotional_subset = emotional_train_data[selected_columns]
```

```python
emotional_subset.head()
```

```python
emotional_subset = emotional_subset.rename(columns={"사람문장1": "text", "감정_대분류": "emotion"})
```

```python
emotional_subset.head()
```

```python
selected_columns = ["text", "emotion"]
neutral_subset = neutral_train_data[selected_columns]
```

```python
neutral_subset.head()
```

```python
#감정데이터셋 비중이 압도적으로 높으므로 개수를 줄임
emotional_subset = emotional_subset.sample(n=3657, random_state=42)
```

```python
len(emotional_subset)
```

```python
combined_data = pd.concat([emotional_subset, neutral_subset], ignore_index=True)
```

```python
combined_data
```

```python
shuffled_df = combined_data.sample(frac=1, random_state=42)
```

```python
shuffled_df
```

```python
shuffled_df.to_csv('/content/drive/MyDrive/졸프/trainingData.csv', index=False)
```
